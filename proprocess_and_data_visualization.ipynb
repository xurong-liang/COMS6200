```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

dataset_dir = "./dataset/"
```

## Read 3 raw dataset and combine into one

```python
# raw metasploitable-2 dataset
raw_meta = pd.read_csv(os.path.join(dataset_dir, "metasploitable-2.csv"))
display(raw_meta)

# raw normal data dataset
raw_normal = pd.read_csv(os.path.join(dataset_dir, "Normal_data.csv"))
display(raw_normal)

# raw OVS dataset
raw_OVS = pd.read_csv(os.path.join(dataset_dir, "OVS.csv"))
display(raw_OVS)
```

```python
# add meta or OVS tag to each label in raw_meta and raw_OVS
for cat in ["meta", 'OVS']:
    locals()[f"raw_{cat}"]["Label"] = f"{cat.lower()}_" + locals()[f"raw_{cat}"]["Label"]
```

```python
# concat entire to one big datset
raw_all = pd.concat([raw_normal, raw_meta, raw_OVS], ignore_index=True)
raw_all.to_csv(os.path.join(dataset_dir, "raw_entire_dataset.csv"), index=False)
```

## Import entire raw dataset for next stage of preprocessing and data exploration

```python
raw_all = pd.read_csv(os.path.join(dataset_dir, "raw_entire_dataset.csv"))
```

```python
raw_all
```

```python
# column alias
col_alias = {name: count for count, name in enumerate(raw_all.columns)}
print(col_alias)
```

```python
# fill oruiginal dimension of dataset
row_num, col_num = raw_all.shape
```

## Drop nan cells, remove attributes with 1 distinct values, remove overly-sparse attributes

```python
# find nan cells 
all(raw_all.isna().sum() == 0)
```

```python
# find spread of distinct counts of each columns and number of classes
distinct_count = raw_all.nunique(axis=0)
distinct_count = dict(sorted(distinct_count.items(), key=lambda item: item[1], reverse=True))

print(f"Number of classes {distinct_count['Label']}")

# delete Label in distinct count so it does not get removed in sparsity pruning
del distinct_count["Label"]
print(distinct_count)
```

```python
# visualize the spread of distinct values of each column
plt.figure(figsize=(20, 5))
plt.title("Number of distinct values per column")
plt.ylabel("count")
plt.xlabel("column name (alias)")
# use horizontal line to reveal columns 
plt.axhline(y=1, c='r', linestyle='-')

x = [count for count in range(len(distinct_count.keys()))]
y = distinct_count.values()
plt.bar(x, y)
plt.xticks(ticks=x, labels=[col_alias[name] for name in distinct_count.keys()])
```

```python
# remove columns that has unique values < 2
for col, count in distinct_count.items():
    if count < 2:
        del raw_all[col]
        print(f'{col} removed')
```

```python
raw_all.shape
```

```python
# find sparsity rate of each non-binary attribute
distinct_count = raw_all.nunique(axis=0)
distinct_count = dict(sorted(distinct_count.items(), key=lambda item: item[1], reverse=True))
sparsity_rates = {}

binary_cols = []

for col, count in distinct_count.items():
    if count > 2 :
        sparsity_rates[col] = (raw_all[col] == 0).sum() / row_num 
    elif count == 2 and set(raw_all[col].unique()) == {0, 1}:
        # binary column, no need to compute sparisty
        
        binary_cols.append(col)
    else:
        # something wrong
        print(col, count)
        break
sparsity_rates = dict(sorted(sparsity_rates.items(), key=lambda item: item[1], reverse=True))   
sparsity_rates
```

```python
binary_cols
```

```python
# plot sparisty rate of non-binary attribute
plt.figure(figsize=(20, 5))
plt.title("Sparisty rates of non-binary attributes")
plt.xlabel("Column name (alias)")
plt.ylabel("Sparisty rate/%")
# candidate pruning thresholds as 60% and 70%
plt.axhline(y=70, c="r", linestyle="-")

plt.axhline(y=60, c="m", linestyle="-")


x = [count for count in range(len(sparsity_rates.keys()))]
y = [100 * rate for rate in sparsity_rates.values()]
plt.bar(x, y)
plt.xticks(x, labels=[col_alias[name] for name in sparsity_rates.keys()])
```

```python
# find number of attrbutes left after pruning
binary_count = len(binary_cols)
count = binary_count
for rate in sparsity_rates.values():
    if rate <= .7:
        count += 1
count
```

```python
# remove overly sparse attributes
for col, rate in sparsity_rates.items():
    if rate > .7:
        del raw_all[col]
        print(f"{col} deleted")
```

## Formatize column names

```python
import re
```

```python
# formatize column names

# convert column names as lowercase, replace space and / as _
new_cols = raw_all.columns.map(lambda x: x.lower()).map(lambda x: re.sub(' ', '_', x)).map(lambda x: re.sub('/', '_', x))
raw_all.rename(columns={key: new_cols[idx] for idx, key in enumerate(raw_all.columns)}, inplace=True)

raw_all.rename(columns={"label": "text_label"}, inplace=True)

raw_all
```

## convert features to numerical values

```python
# timestamp
raw_all["timestamp"] = pd.to_datetime(raw_all["timestamp"], infer_datetime_format=True).astype(int) / 10**9
```

```python
raw_all
```

```python
import ipaddress
```

```python
# convert ip address to integer

for ip in ["src_ip", "dst_ip"]:
    raw_all[ip] = raw_all[ip].map(lambda x: int(ipaddress.ip_address(x)))

raw_all
```

```python
# change flow id to numerical values
raw_all["flow_id"] = pd.factorize(raw_all["flow_id"])[0]

raw_all
```

## Add ordinal labels

```python
raw_all["ordinal_label"] = pd.factorize(raw_all["text_label"])[0]
```
